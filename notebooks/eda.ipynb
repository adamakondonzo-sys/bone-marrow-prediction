```python
# Cellule 1 — Imports et chargement des données
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo

# Chargement du dataset
dataset = fetch_ucirepo(id=565)
df = pd.concat([dataset.data.features, dataset.data.targets], axis=1)

print("Shape du dataset :", df.shape)
print("\nAperçu des données :")
df.head()
```

```python
# Cellule 2 — Structure et types de données
print("Dimensions :", df.shape)
print("\nTypes de données :")
print(df.dtypes)
print("\nStatistiques descriptives :")
df.describe()
```

```python
# Cellule 3 — Analyse des valeurs manquantes
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Valeurs manquantes': missing,
    'Pourcentage (%)': missing_pct.round(2)
}).query('`Valeurs manquantes` > 0').sort_values('Pourcentage (%)', ascending=False)

print(missing_df)

# Visualisation
import missingno as msno
msno.matrix(df)
plt.show()
```

```python
# Cellule 4 — Traitement des valeurs manquantes
from sklearn.impute import SimpleImputer, KNNImputer

# Colonnes avec > 5% manquants → KNN + flag
cols_knn = ['extcGvHD', 'CMVstatus', 'RecipientCMV']

for col in cols_knn:
    df[col + '_was_missing'] = df[col].isnull().astype(int)

imp_knn = KNNImputer(n_neighbors=5)
df[cols_knn] = imp_knn.fit_transform(df[cols_knn])

# Colonnes avec < 5% manquants → médiane
cols_median = ['CD3dCD34', 'CD3dkgx10d8', 'DonorCMV', 
               'RecipientRh', 'Rbodymass', 'ABOmatch', 
               'RecipientABO', 'Allele', 'Antigen']

imp_median = SimpleImputer(strategy='median')
df[cols_median] = imp_median.fit_transform(df[cols_median])

# Vérification
print("Valeurs manquantes restantes :", df.isnull().sum().sum())
```

```python
# Cellule 5 — Détection des valeurs aberrantes
num_cols = df.select_dtypes(include=np.number).columns

# Boxplots
df[num_cols].plot(
    kind='box',
    subplots=True,
    layout=(6, 7),
    figsize=(20, 15)
)
plt.suptitle('Détection des valeurs aberrantes', y=1.02)
plt.tight_layout()
plt.show()

# Compter les outliers par colonne avec IQR
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

outliers = ((df[num_cols] < Q1 - 1.5*IQR) | (df[num_cols] > Q3 + 1.5*IQR)).sum()
print("\nNombre d'outliers par colonne :")
print(outliers[outliers > 0].sort_values(ascending=False))
```

```python
# Cellule 6 — Traitement des outliers (winsorisation)
from scipy.stats import mstats

# Uniquement les variables continues médicales
cols_to_winsorize = ['CD34kgx10d6', 'Rbodymass', 'CD3dkgx10d8', 
                     'CD3dCD34', 'Donorage', 'Recipientage']

df_before = df[cols_to_winsorize].copy()

for col in cols_to_winsorize:
    df[col] = mstats.winsorize(df[col], limits=[0.01, 0.01])

print("Winsorisation appliquée sur :", cols_to_winsorize)

# Comparaison avant/après
fig, axes = plt.subplots(2, len(cols_to_winsorize), figsize=(18, 8))

for i, col in enumerate(cols_to_winsorize):
    df_before[col].plot(kind='box', ax=axes[0, i], title=col)
    axes[0, i].set_xlabel('Avant')
    df[col].plot(kind='box', ax=axes[1, i])
    axes[1, i].set_xlabel('Après')

plt.suptitle('Avant vs Après Winsorisation')
plt.tight_layout()
plt.show()
```

```python
# Cellule 7 — Déséquilibre des classes
target = 'survival_status'

# Distribution
print("Distribution des classes :")
print(df[target].value_counts())
print("\nEn pourcentage :")
print(df[target].value_counts(normalize=True).round(3) * 100)

# Visualisation
df[target].value_counts().plot(
    kind='bar',
    color=['steelblue', 'salmon'],
    title='Distribution des classes',
    ylabel='Nombre de patients',
    xlabel='Survie'
)
plt.xticks([0, 1], ['Survécu (1)', 'Non survécu (0)'], rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Cellule 8 corrigée — Encodage + Split + SMOTE
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder

X = df.drop(columns=['survival_status'])
y = df['survival_status']

# Encoder les colonnes texte en chiffres
le = LabelEncoder()
for col in X.select_dtypes(include='object').columns:
    print(f"Encodage de : {col}")
    X[col] = le.fit_transform(X[col].astype(str))

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("\nAvant SMOTE :")
print(y_train.value_counts())

# SMOTE uniquement sur le train
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

print("\nAprès SMOTE :")
print(pd.Series(y_train_res).value_counts())
```

```python
# Cellule 9 — Corrélations
num_cols = X.select_dtypes(include=np.number).columns
corr_matrix = X[num_cols].corr()

# Heatmap
plt.figure(figsize=(16, 12))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=False,
    cmap='coolwarm',
    vmin=-1, vmax=1,
    linewidths=0.5
)
plt.title('Matrice de corrélation')
plt.tight_layout()
plt.show()

# Paires fortement corrélées
threshold = 0.85
high_corr_pairs = []
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        val = abs(corr_matrix.iloc[i, j])
        if val > threshold:
            high_corr_pairs.append({
                'Feature 1': corr_matrix.columns[i],
                'Feature 2': corr_matrix.columns[j],
                'Corrélation': round(val, 3)
            })

high_corr_df = pd.DataFrame(high_corr_pairs).sort_values('Corrélation', ascending=False)
print("\nPaires fortement corrélées (|r| > 0.85) :")
print(high_corr_df)
```

```python
# Cellule 10 — Suppression des features redondantes
cols_to_drop = [
    'time_to_aGvHD_III_IV',
    'HLAgrI',
    'RecipientCMV_was_missing',
    'Recipientageint',
    'Allele'
]

X = X.drop(columns=cols_to_drop)
X_train = X_train.drop(columns=cols_to_drop)
X_test = X_test.drop(columns=cols_to_drop)
X_train_res = pd.DataFrame(X_train_res, columns=X_train.columns.tolist() + ['Allele']).drop(columns=cols_to_drop, errors='ignore')

print("Dimensions après suppression :", X.shape)
print("Colonnes supprimées :", cols_to_drop)
```
